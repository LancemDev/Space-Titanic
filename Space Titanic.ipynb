```python
#Creating a model
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
x_train=df_train[["Age"]]
y_train=df_train[["Transported"]]
x_test=df_test[["Age"]]
dtc.fit(x_train,y_train)
y_hats=dtc.predict(x_test)
```

```python
#Creating a submission csv file
#first we need to concatenate the predicted column
df_test['Transported']=y_hats
df_submission=pd.merge(df_test,df_test[["Transported"]])
```
